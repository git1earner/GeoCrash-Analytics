```python
import pandas as pd

file_path = "../data/US_Accidents_March23.csv"

df = pd.read_csv(
    file_path,
    skiprows=lambda x: x > 0 and x % 50 != 0
)

print("Rows:", len(df))
print("Columns:", len(df.columns))
```

```python
df.columns.tolist()
```

```python
missing = df.isnull().sum()

missing.sort_values(
    ascending=False
).head(20)
```

```python
df["State"].value_counts().head(10)
```

```python
import matplotlib.pyplot as plt

state_counts = df["State"].value_counts()

plt.figure(figsize=(12,6))

state_counts.head(10).plot(kind="bar")

plt.title("Top 10 States by Accident Count")
plt.xlabel("State")
plt.ylabel("Accident Count")

plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(10,8))

plt.scatter(
    df["Start_Lng"],
    df["Start_Lat"],
    s=1,
    alpha=0.3
)

plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("US Accident Locations")

plt.savefig(
    "../images/us_accident_locations.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```
